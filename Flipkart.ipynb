```python
import requests
```

```python
data = requests.get('https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=0&as-type=HISTORY')
```

```python
from bs4 import BeautifulSoup
```

```python
flip = BeautifulSoup(data.text,'lxml')
```

```python
flip
```

```python
names = []
for var in flip.find_all('a',attrs={'class':'_2cLu-l'}):
    names.append(var.text)
```

```python
names
```

```python
prices = []
for var in flip.find_all('div',attrs={'class':'_1vC4OE'}):
    prices.append(var.text)
```

```python
prices
```

```python
ratings = []
for var in flip.find_all('div',attrs={'class':'hGSR34'}):
    ratings.append(var.text)
```

```python
ratings
```

```python
len(names) == len(prices)
```

```python
len(prices)
```

```python
len(names)
```

```python
len(ratings)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
        'name' : names,
        'price' : prices[:40],
        'ratings' : ratings[:40]
})
#pd.to_csv("one.csv",df)
```

```python
df
```

```python
df.to_csv('flip.csv',index=False)
```
