```python
soup.select('link')
```

```python
soup.select('a')
```

```python
for var in soup.select('a'):
    print(var.text)
```

```python
for var in soup.select('a'):
    try:
        print(var['href'])
    except:
        pass
        
```

```python
for var in soup.select('div'):
    print(var.text)
```

```python
for var in soup.select('.toc'):
    print(var.text)
```

```python
soup.find_all('a',attrs={'class':'mw-redirect'})
```

```python
for var in soup.find_all('img'):
    print(var['src'])
```

```python
for var in soup.find_all('div',attrs={'class':['div-col','columns','column-width']}):
    print(var.text)
```
